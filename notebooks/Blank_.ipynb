## first observations
Further attention needed for RELATIVE_PERCENT_DIFFERENCE and PRECISION_STATUS. how many missing values are there in this dataset.

```python
# Check for missing values
missing_values = df.isnull().sum()
print("\nPercentage of missing values:")
for name, values in missing_values.items():
    if values > 0:
        print(name, ":", round((values / len(df) * 100),1), '%')

# removing columns with no null values
sorted_missing_values = missing_values[missing_values > 0].sort_values(ascending=False)
import matplotlib.pyplot as plt
plt.bar(data= sorted_missing_values, x=sorted_missing_values.index, height=sorted_missing_values.values)
plt.title('Missing Values per Column')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#create clean_df with columns RELATIVE_PERCENT_DIFFERENCE and PRECISION_STATUS dropped
clean_df = df.drop(columns=['RELATIVE_PERCENT_DIFFERENCE', 'PRECISION_STATUS'])
```

```python
import seaborn as sns

import matplotlib.pyplot as plt
# Select numeric columns (int64 and float64)
numeric_cols = clean_df.select_dtypes(include=['int64', 'float64']).columns

# Generate a separate boxplot for each numeric column
for col in numeric_cols:
    if col in ['SCHEME_VERSION']:
        continue  # Skip unnecessary columns
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=clean_df, y=col)
    plt.title(f'Boxplot of {col}')
    plt.show()
```

## looking at differences in values of blank samples

```python
#create an overview of all unqiue values in ANALYTICAL_TYPE and their counts
print(clean_df['ANALYTICAL_TYPE'].value_counts())
```

```python
#convert ANALYSED_DATE to datetime format
clean_df['ANALYSED_DATE'] = pd.to_datetime(clean_df['ANALYSED_DATE'])  
clean_df.head()


#show the date range of the dataset
print("Date range:", clean_df['ANALYSED_DATE'].min(), "to", clean_df['ANALYSED_DATE'].max())
```

```python
# visualise NUMERIC_FINAL_VALUE and	INTERNAL_TARGET_VALUE	 together in a scatter plot filter on analytical type = 'Blank' and one on analytical type = 'Standard' with ANALYSED_DATE as the x axis
# Filter the DataFrame for 'Blank' and 'Standard' analytical types
# filter to data from 2021 only 
    
blank_df = clean_df[(clean_df['ANALYTICAL_TYPE'] == 'Blank') & (clean_df['ANALYSED_DATE'].dt.year.isin([2021]))]
standard_df = clean_df[(clean_df['ANALYTICAL_TYPE'] == 'Standard') & (clean_df['ANALYSED_DATE'].dt.year.isin([2021]))]
# Create scatter plots for both types
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)    
sns.scatterplot(data=blank_df, x='ANALYSED_DATE', y='NUMERIC_FINAL_VALUE')
sns.scatterplot(data=blank_df, x='ANALYSED_DATE', y='INTERNAL_TARGET_VALUE')
plt.title('Blank Analytical Type (in 2021)')
plt.xlabel('ANALYSED_DATE')
plt.ylabel('Value')
#show the legend outside of the graph
plt.legend(['NUMERIC_FINAL_VALUE', 'INTERNAL_TARGET_VALUE'], loc='upper right')


# standar values in a plot next to the blank plot
plt.subplot(1, 2, 2)
sns.scatterplot(data=standard_df, x='ANALYSED_DATE', y='NUMERIC_FINAL_VALUE')
sns.scatterplot(data=standard_df, x='ANALYSED_DATE', y='INTERNAL_TARGET_VALUE')
plt.title('Standard Analytical Type (in 2021)')
plt.xlabel('ANALYSED_DATE')
plt.ylabel('Value')

plt.show()
```

```python
# show the distribution of NUMERIC_FINAL_VALUE for analytical type = 'Blank' and 'Standard' in a histogram
# make binranges with a size of 100000
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)        
sns.histplot(data=blank_df, x='NUMERIC_FINAL_VALUE', bins = 100)
plt.title('Blank Analytical Type (in 2021)')
plt.xlabel('NUMERIC_FINAL_VALUE')
plt.ylabel('Count')
plt.subplot(1, 2, 2)
sns.histplot(data=standard_df, x='NUMERIC_FINAL_VALUE', bins = 100)
plt.title('Standard Analytical Type (in 2021)')
plt.xlabel('NUMERIC_FINAL_VALUE')
plt.ylabel('Count')
plt.show()
    

```
